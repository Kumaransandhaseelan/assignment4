```python
n=int(input("Enter number: "))

def prime(n):
    if n>1:
        
        for i in range(2,int(n/2)+1):
            if n%i==0:
                return(str(n)+" is not a prime number")
                break
        return(str(n)+"  is a Prime Number")
    else:
        print(str(n)+" is a Prime Number")
 
 
print(prime(n))
```
